```python
df = df.rename(columns = {"Mat" : "Matches", "NO" : "Not_outs", "HS" : "Highest_inns_score", "Ave" : "Batting_average" , "BF" : "Balls_faced", "SR" : "Batting_strike_rate" })
```

```python
# Splitting the "Span" column of into two parts thorugh "-"

df["Span"].str.split(pat = " - ")
```

```python
# Splitting the "Span" column of into two parts thorugh "-"
# This would be our Rookie_Year for the cricketer  

df["Span"].str.split(pat = "-").str[0]
```

```python
# Splitting the "Span" column of into two parts thorugh "-"
# This would be our Final_Year for the cricketer  

df["Span"].str.split(pat = "-").str[1]
```

```python
# Splitting the "Span" column into two parts through "-"
# Here we have created 2 columns for the player as their Rookie_Year and Final_Year

df["Rookie_Year"] = df["Span"].str.split(pat = "-").str[0]
df["Final_Year"] = df["Span"].str.split(pat = "-").str[1]
```

```python
# Droping the "Span" column because it is not needed anymore 

df = df.drop(["Span"], axis = 1)
```

```python
df.head()
```

```python
#QUESTION -> Split up the country from the player
```

```python
df["Player"].str.split(pat = "(")
```

```python
# Spliting the country and putting it in another column

df["Country"] = df["Player"].str.split(pat = "(").str[1]
df["Country"] = df["Country"].str.split(pat = ")").str[0]
```

```python
df["Country"]
```

```python
# Spliting the Player column for only the names of the player

df["Player"] = df["Player"].str.split(pat = "(").str[0]
```

```python
df.head() # We have split the Player and Country name into two different columns
```

```python
df["Highest_inns_score"] = df["Highest_inns_score"].str.split(pat = "*").str[0]
```

```python
df["Highest_inns_score"] = df["Highest_inns_score"].astype("int")
```

```python
df["Matches"] = df["Matches"].str.split(pat = "*").str[0]
```

```python
df["Matches"] = df["Matches"].astype("int")
```

```python
# QUESTION 2 -> Build out career length column 
# Rookie Year - Final Year 
```

```python
df["career_length"] = df["Final_Year"] - df["Rookie_Year"] 
```

```python
df # Here we have calculated the career length of each player
```

```python
# QUESTION 3 -> Cricketers in this df what is the Average Carrer length
```

```python
df["career_length"].mean()
```

```python
# QUESTION 4 -> Average Batting_strike_rate for cricketers who played over 10 years
```

```python
df[df["career_length"] > 10]["Batting_strike_rate"].mean()
```

```python
# QUESTION 5 -> Find the number of cricketers who played before 1960 
```

```python
df[df["Rookie_Year"] < 1960]["Player"].count()
```

```python
# QUESTION 6 -> Max highest inns score by country
```

```python
df.groupby("Country")["Highest_inns_score"].max().to_frame("Highest_inns_score").reset_index().sort_values("Highest_inns_score", ascending = False)
```
